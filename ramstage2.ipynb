```python
import pandas as pd
pd.set_option('display.precision',15)
df_npa = pd.read_csv(r'C:\Users\KC\Desktop\EssilorWP\NPA_Forstage2.csv')
df_macola = pd.read_csv(r"C:\Users\KC\Desktop\WhitespaceProject\Macola.csv")
df_npa.set_index('Unnamed: 0', inplace=True)
df_macola['Phone'] = df_macola['Phone'].astype(str)          
df_macola['Phone'] = df_macola['Phone'].str.replace('.0', ' ')
df_macola['Practice Phone'] = df_macola['Practice Phone'].astype(str)          
df_macola['Practice Phone'] = df_macola['Practice Phone'].str.replace('.0', ' ')
df_npa['Provider Business Mailing Address Telephone Number'] = df_npa['Provider Business Mailing Address Telephone Number'].astype(str)          
df_npa['Provider Business Mailing Address Telephone Number'] = df_npa['Provider Business Mailing Address Telephone Number'].str.replace('.0', ' ')
df_npa['Provider Business Practice Location Address Telephone Number'] =df_npa['Provider Business Practice Location Address Telephone Number'].astype(str)          
df_npa['Provider Business Practice Location Address Telephone Number'] =df_npa['Provider Business Practice Location Address Telephone Number'].str.replace('.0', ' ')
```

```python
directmatch =  df_npa.index.tolist()
df_npa.drop(['check1'],axis=1)
df_npa.drop(['check2'],axis=1)
```

```python
cols_practice_macola = ['Practice Street Address 1', 'Practice Street Address 2','Practice City']
df_macola['check1'] = df_macola[cols_practice_macola].apply(lambda x: ','.join(x.dropna()), axis=1)
cols_practice_macola = ['Shipping Address Line 1', 'Shipping Address Line 2', 'Shipping Address Line 3', 'Shipping City']
df_macola['check2'] = df_macola[cols_practice_macola].apply(lambda x: ','.join(x.dropna()), axis=1)
```

```python
cols_practice_npa = ['Provider First Line Business Practice Location Address', 'Provider Second Line Business Practice Location Address','Provider Business Practice Location Address City Name']
df_npa['check1'] = df_npa[cols_practice_npa].apply(lambda x: ','.join(x.dropna()), axis=1)
cols_practice_npa = ['Provider First Line Business Mailing Address','Provider Second Line Business Mailing Address','Provider Business Mailing Address City Name']
df_npa['check2'] = df_npa[cols_practice_npa].apply(lambda x: ','.join(x.dropna()), axis=1)
```

```python
df_macola['flag'] = None
df_macola['flag1'] = None
index_r1 = []
macola_id_l1 = []
pon = []
df_pp = pd.DataFrame()
address = df_macola['check1'].tolist()
df_macola['PP Old'] = df_macola['Practice Phone']
for item in directmatch:
    a=df_npa.loc[df_npa.index == item, 'check1'].iloc[0]
    if a in address:
        b=df_macola.loc[df_macola['check1'] == a, 'Practice Phone'].values
        if len(b)>0:
            #macola_id=df_macola.loc[df_macola['Practice Phone'] == a, 'Macola/Company ID'].values
            indexpp = df_macola['check1'][df_macola['check1'] == a].index.tolist()
            c=df_npa.loc[df_npa.index == item, 'Provider Business Practice Location Address Telephone Number'].iloc[0]
            d=df_npa.loc[df_npa.index == item, 'Provider Other Organization Name'].values
            pn=df_macola.loc[df_macola['check1'] == a, 'Practice Name'].values
            if b[0]==c:
                macola_id_l1.append(item)
            else:
                  
                df_macola.loc[indexpp[0],'Practice Phone'] = c
                df_macola.loc[indexpp[0],'flag'] = 'PP-NA'
                df_pp = df_pp.append([df_macola[df_macola.index==indexpp[0]]])
                index_r1.append(item)
                if len(d)>0 & len(pn)>=0:
                    if d[0] != pn[0]:
                        df_pp.loc[indexpp[0],'Practice Name'] = d[0]
                        
                        df_pp.loc[indexpp[0],'flag1'] = 'PN-NA'
                        pon.append(item)
            

df_pp.shape
```

```python
df_pp = df_pp.drop(['check1'], axis = 1)
df_pp = df_pp.drop(['check2'], axis = 1) 
df_pp.to_csv(r'C:\Users\KC\Desktop\macola\straightPP-NPupdate1.csv')
```

```python
df_macola['flag2'] = None
df_macola['flag3'] = None
index_r2 = []
macola_id_l2 = []
pon2 = []
df_bp = pd.DataFrame()
address = df_macola['check2'].tolist()
df_macola['BP Old'] = df_macola['Phone']
for item in directmatch:
    a=df_npa.loc[df_npa.index == item, 'check2'].iloc[0]
    if a in address:
        b=df_macola.loc[df_macola['check2'] == a, 'Phone'].values
        if len(b)>0:
            #macola_id=df_macola.loc[df_macola['Practice Phone'] == a, 'Macola/Company ID'].iloc[0]
            indexpp2 = df_macola['check2'][df_macola['check2'] == a].index.tolist()
            c=df_npa.loc[df_npa.index == item, 'Provider Business Mailing Address Telephone Number'].iloc[0]
            d=df_npa.loc[df_npa.index == item, 'Provider Organization Name (Legal Business Name)'].values
            pn=df_macola.loc[df_macola['check2'] == a, 'Account Name'].values
            if b[0]==c:
                #index_r2.append(item)
                macola_id_l1.append(item)
            else:  
                df_macola.loc[indexpp2[0],'Phone'] = c
                df_macola.loc[indexpp2[0],'flag2'] = 'BP-NA'
                df_bp = df_bp.append([df_macola[df_macola.index==indexpp2[0]]])
                index_r2.append(item)
                if len(d)>0 & len(pn)>=0:
                    if d[0] != pn[0]:
                        df_macola.loc[indexpp2[0],'Practice Name'] = d[0]
                        df_macola.loc[indexpp2[0],'flag3'] = 'BN-NA'
                        pon2.append(item)
            
df_bp.shape
```
