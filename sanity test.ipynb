```python
import jellyfish
from pyjarowinkler import distance
import re

cols1 = ['Shipping Address Line 1', 'Shipping Address Line 2', 'Shipping Address Line 3', 'Shipping City']
cols = ['Shipping Street','Shipping City']
df_macola['check1'] = df_macola[cols1].apply(lambda x: ','.join(x.dropna()), axis=1)
df_macola['check2'] = df_macola[cols].apply(lambda x: ','.join(x.dropna()), axis=1)
cols_npa_stage1 = ['Provider First Line Business Mailing Address', 'Provider Second Line Business Mailing Address', 'Provider Business Mailing Address City Name']
df_npa_stage1['check1'] = df_npa_stage1[cols_npa_stage1].apply(lambda y: ','.join(y.dropna()), axis=1)
index_jf = []
for index,row in df_npa_stage1.iterrows():
    for index1,row1 in df_macola.iterrows():
        check1 = re.sub('[^a-zA-Z0-9]', '', row['check1'])
        check2 = re.sub('[^a-zA-Z0-9]', '', row1['check1'])
        #check3 = re.sub('[^a-zA-Z0-9]', '', row['fullname'])
        #check4 = re.sub('[^a-zA-Z0-9]', '', row1['fullname'])
        if (jellyfish.jaro_winkler(check1,check2)>.50) & (str(check1[0:8]) == str(check2[0:8])):
            if len(re.findall(r'\d+', check1))==2 & len(re.findall(r'\d+', check2))==2:
                if (re.findall(r'\d+', check1)[1] == re.findall(r'\d+', check2)[1]):
                    index_jf.append(index)
                    print(re.findall(r'\d+', check1)[1])
                    print(re.findall(r'\d+', check2)[1])
                    f = open(r"C:\Users\KC\Desktop\EssilorWP\sanitytest3.txt", 'a')
                    f.write(str(index) + '\n')
                    f.close 
            if len(re.findall(r'\d+', check1))==1 & len(re.findall(r'\d+', check2))==1:
                if (re.findall(r'\d+', check1)[0] == re.findall(r'\d+', check2)[0]):
                    index_jf.append(index)
                    print("///////////")
                    print(re.findall(r'\d+', check1)[0])
                    print(re.findall(r'\d+', check2)[0])
                    f = open(r"C:\Users\KC\Desktop\EssilorWP\sanitytest3.txt", 'a')
                    f.write(str(index) + '\n')
                    f.close 
print(len(index_jf))
```
